```python
N_tot = 1000
N_men = N_tot // 2
N_women = N_tot // 2
num_groups_men = 4
num_groups_women = 4

# Generate ranks
s_men = np.random.uniform(low=1, high=10, size=N_men)
s_women = np.random.uniform(low=1, high=10, size=N_women)

# Generate groups
g_men = np.random.randint(0, num_groups_men, size=N_men)
G_m = np.zeros((N_men, num_groups_men))
for man_id, group_id in enumerate(g_men):
    G_m[man_id, group_id] = 1
    
# Generate group preferences
group_pref = np.abs(np.random.normal(0, 1))
group_disc = np.abs(np.random.normal(0, 1))

# Theta[i, j] - Preference of woman in group i towards a man in group j
Theta = np.zeros((num_groups_women, num_groups_men))
for i in range(num_groups_women):
    Theta[num_groups_women-1, i] = -group_disc
    Theta[i, i] = group_pref # overwrite for last group
```

```python
def generate_network(N, s, G, theta, beta):
    
    ranks = s + np.matmul(G, theta)
    
    Z = 0
    scaled_energy = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            energy_ij = 0.5 * np.power(ranks[i] - ranks[j] - 1, 2)
            scaled_energy[i, j] =  np.exp(-beta * energy_ij)
            Z += scaled_energy[i, j]
    c = 10 * N / Z
    
    A = np.zeros((N, N))
    for i in range(N):
        for j in range(N):
            if i == j:
                continue
            A[i, j] += np.random.poisson(c * scaled_energy[i, j])
            
    return A
```

```python
beta = 0.5

A_t = []
A = np.zeros((N_men, N_men))

for i in range(num_groups_women):
    A_i = generate_network(N_men, s_men, G_m, Theta[i], beta)
    A_t.append(A_i)
    A += A_i
```

```python
s_est = sr.SpringRank(A)
```

### Extract hierarchy with groups

```python
def construct_system(A, A_t, G, l_a, l_b):
    """
    Construct the linear system to solve for
    
    Arguments:
        A: The full network
        A_t: List of networks where A_t[i] is the network decided by arbiters in group i
        G: The group identification matrix for suitors
        l_a: Regularization for full network
        l_b: List, regularization for each set of arbiters
    
    Returns:
        LHS: The left hand side of the linear system
        RHS: The right hand side of the linear system
    """
    
    # Construct full Laplacian
    N = A.shape[0]
    N_groups_arbiters = len(A_t)
    N_groups_suitors = G.shape[1]
    k_in = np.sum(A, 0)
    k_out = np.sum(A, 1)
    L = np.diag(k_out + k_in) - (A + A.T)
    
    # Set regularization
    if l_a is None:
        l_a = 0
    if l_b is None:
        l_b = [0] * N_groups_arbiters

    # Construct group Laplacians
    L_t = []
    k_t_in = []
    k_t_out = []
    for A_r in A_t:
        N_r = A_r.shape[0]
        k_r_in = np.sum(A_r, 0)
        k_r_out = np.sum(A_r, 1)
        L_r = np.diag(k_r_out + k_r_in) - (A_r + A_r.T)

        L_t.append(L_r)
        k_t_in.append(k_r_in)
        k_t_out.append(k_r_out)

    # Row 1
    LHS = L + l_a*np.eye(N)
    for L_r in L_t:
        LHS = np.concatenate((LHS, np.matmul(L_r, G)), axis=1)
    RHS = k_out - k_in

    filler_matrix = np.zeros((N_groups_suitors, N_groups_suitors))

    # Row n
    for i, L_r in enumerate(L_t):
        compressed = np.matmul(G.T, L_r)
        row_n = np.copy(compressed)
        block = np.matmul(compressed, G) + l_b[i]*np.eye(N_groups_suitors)
        for j in range(N_groups_arbiters):
            if i == j:
                row_n = np.concatenate((row_n, block), axis=1)
            else:
                row_n = np.concatenate((row_n, filler_matrix), axis=1)
        LHS = np.concatenate((LHS, row_n), axis=0)
        RHS = np.concatenate((RHS, np.matmul(G.T, k_t_out[i] - k_t_in[i])), axis=0)
        
    return LHS, RHS

def extract_hierarchy(A, A_t, G, l_a=None, l_b=None):
    """
    Solve the linear system and extract the ranks and preferences
    
    Arguments:
        A: The full network
        A_t: List of networks where A_t[i] is the network decided by arbiters in group i
        G: The group identification matrix for suitors
        l_a: Regularization for full network
        l_b: List, regularization for each set of arbiters
        
    Returns:
        s_est: Estimated individual scores
        Theta_est: Estimated matrix of preferences
            Theta_est[i, j] - Preference of woman in group i towards a man in group j
    """
    
    N_suitors = A.shape[0]
    N_groups_arbiters = len(A_t)
    N_groups_suitors = G.shape[1]
    
    LHS, RHS = construct_system(A, A_t, G_m, l_a, l_b)
    output = spLA.spsolve(LHS, RHS)
    x = output[0]
    
    s_est = x[:N_suitors]
    Theta_est = np.zeros((N_groups_arbiters, N_groups_suitors))
    for r in range(N_groups_arbiters):
        start_idx = N_men + (r*N_groups_suitors)
        end_idx = N_men + (r*N_groups_suitors) + N_groups_suitors
        theta_r = x[start_idx:end_idx]
        for s, val in enumerate(theta_r):
            Theta_est[r, s] = val
            
    return s_est, Theta_est
```
